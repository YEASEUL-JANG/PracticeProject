```python
N = 5
graph = [[0,1,0,1,1],
         [1,0,1,1,0],
         [0,1,0,0,0],
         [1,1,0,0,0],
         [1,0,0,0,0]]
dist = [2,3,4,3]
parents = [0] * (N+1)

for i in range(N+1):
    parents[i] = i
    
def find(parents,a):
    if parents[a]!=a:
        parents[a] = find(parents,parents[a])
    return parents[a]

def union(parents,a,b):
    par_a = find(parents,a)
    par_b = find(parents,b)
    if par_a<par_b:
        parents[par_b] = par_a
    else:
        parents[par_a] = par_b

for i in range(N):
    for j in range(N):
        if graph[i][j]==1:
            union(parents,i,j)
result = True
for i in range(len(dist)-1):
    if find(parents,dist[i])!=find(parents,dist[i+1]):
        result=False
        break
print(result)
        
    
```

```python
N = 7 #집의 수
M = 11 #도로의 수
graph = []
for i in range(M):
    a,b,cost = map(int,input().split())
    graph.append((cost,a,b))
graph.sort()

parents = [0]*(N+1)
for i in range(1,N+1):
    parents[i] = i

def find(parents,a):
    if parents[a]!=a:
        parents[a] = find(parents,parents[a])
    return parents[a]
    
def union(parents,a,b):
    par_a = find(parents,a)
    par_b = find(parents,b)
    if par_a<par_b:
        parents[par_b]=par_a
    else:
        parents[par_a]=par_b
        
totalcost = 0
savecost = 0
for g in graph:
    cost,a,b = g
    totalcost+=cost
    if find(parents,a)!= find(parents,b):
        union(parents,a,b)
        savecost+=cost
print(totalcost-savecost)
```

```python
N = int(input())
unit = []
for i in range(N):
    x,y,z = map(int,input().split())
    unit.append((x,y,z,i))
graph = []
for i in range(3):
    unit.sort(key=lambda x:x[i])
    for j in range(1,N):
        a,b,c,d = unit[j-1]
        e,f,g,h = unit[j]
        cost = min(abs(a-e), abs(b-f),abs(c-g))
        graph.append((cost,d,h))
graph.sort()

parents = [i for i in range(N)]

def find(parents,a):
    if parents[a] != a:
        parents[a] = find(parents,parents[a])
    return parents[a]

def union(parents,a,b):
    par_a = find(parents,a)
    par_b = find(parents,b)
    if par_a<par_b:
        parents[par_b]=par_a
    else:
        parents[par_a]=par_b
        
def get_answer():
    result = 0
    for g in graph:
        cost,a,b = g
        if find(parents,a)!=find(parents,b):
            union(parents,a,b)
            result+=cost
    return result

print(get_answer())
    
        
```
